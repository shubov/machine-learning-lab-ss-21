# K-Means

```python
import numpy as np
from sklearn.datasets import make_blobs, make_moons, make_circles
import matplotlib.pyplot as plt


from sklearn.decomposition import PCA
%matplotlib inline
```

## Dataset

```python
k = 2
blobs, _ = make_blobs(n_samples=1000, n_features=2, centers=k)
moons, _ = make_moons(n_samples=1000)
noisy_circles = make_circles(n_samples=1000, factor=.5, noise=.05)
```

## Model

```python
from shubov_algorithms import k_means
model = k_means.KMeans()
```

## Testing

```python
for data in [blobs, moons,noisy_circles[0]]:
    centroids = model.do_k_means(data, k, movement_threshold_delta=0)
    clusters = model.compute_clusters(data, centroids)
    
    plt.rcParams['figure.figsize'] = [10, 5]
    for centroid, points in clusters.items():
        points = np.array(points)
        centroid = np.mean(points, axis=0)


        plt.scatter(points[:, 0], points[:, 1], marker='o')
        plt.grid()
        plt.scatter(centroid[0], centroid[1], marker='x', color="red")

    plt.show()
```
